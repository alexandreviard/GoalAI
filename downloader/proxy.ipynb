```python
import requests
import pandas as pd
import concurrent.futures
```

```python
team_url = 'https://fbref.com/en/squads/e2d8892c/Paris-Saint-Germain-Stats'
url_test = 'https://httpbin.org/ip'
wiki = 'https://fr.wikipedia.org/wiki/Wikip%C3%A9dia:Accueil_principal'
valid_proxies
```

```python
response = requests.get('https://free-proxy-list.net/')
table = pd.read_html(response.text)[0]
#table = table[(table['Https'] == 'yes')]


def test(proxy):
    try:
        response = requests.get('https://httpbin.org/ip', proxies={'http':proxy, 'https':proxy}, timeout=10)
        print(response.json(), ' - working')
        return proxy
    except:
        return None

proxies_list = [f"{row['IP Address']}:{row['Port']}" for index, row in table.iterrows()]
proxies_work = []

with concurrent.futures.ThreadPoolExecutor() as executor:
    results = executor.map(test, proxies_list)
    # Filtrer les résultats pour conserver uniquement les proxies valides
    valid_proxies = [result for result in results if result]


proxy_to_csv = []
for proxy in valid_proxies:
    try:
        response = requests.get(team_url, proxies={'http':proxy, 'https':proxy}, timeout=5)
        team_data = pd.read_html(response.text, match="Scores")[0]
        print(team_data)  # Affiche les données extraites
        proxy_to_csv.append(proxy)

        
    except requests.exceptions.ProxyError:
        print(f'Erreur de proxy pour {proxy}')
    except requests.exceptions.ConnectTimeout:
        print(f'Timeout de connexion pour {proxy}')
    except requests.exceptions.SSLError:
        print(f'Erreur SSL pour {proxy}')
    except ValueError as e:
        # Cela se produira si aucune table correspondant à "Scores" n'est trouvée
        print(f'Erreur lors de la lecture HTML pour {proxy}: {e}')
    except Exception as e:
        # Capture toutes les autres exceptions potentielles
        print(f'Erreur inattendue pour {proxy}: {e}')
    finally:
        # Le bloc 'finally' garantit que le code continuera même en cas d'exception
        print(f"Test terminé pour {proxy}\n")

# Créer un DataFrame à partir des proxies valides
df_exist = pd.read_csv('proxy.csv')
df = pd.DataFrame({'Proxy': proxy_to_csv})
pd.concat([df_exist, df]).drop_duplicates(subset="Proxy").to_csv('proxy.csv', index=False)
# Vous pouvez ensuite enregistrer le DataFrame dans un fichier CSV si nécessaire
#df.to_csv("proxy.csv", index=False)
```
